```python
import PyPDF2 as pdf
import textract
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import os
```

```python
rootdir = os.getcwd()+'\RQPS'

files = [ f for f in os.listdir(rootdir) if os.path.isfile(os.path.join(rootdir,f)) ]
for subdir, dirs, files in os.walk(rootdir):
    for fileName in files:
        extension = os.path.splitext(fileName)[-1]
        if(extension==".pdf"):
            file = open(rootdir+'\\'+fileName, mode='rb')
            pdf_reader = pdf.PdfFileReader(file)
            totalPages = pdf_reader.numPages
            currentPage = 0
            text = ""
            while(currentPage < totalPages):
                pdfPage = pdf_reader.getPage(currentPage)
                try:
                    text += pdfPage.extractText()
                except:
                    pass
                currentPage += 1
            if(text == ""):
                text = textract.process(filename=rootdir+'\\'+fileName, method='tesseract', encoding='utf8')
#           Preprocessing
            text = text.strip().lower() # Remove white spaces - to lower case
            tokens = word_tokenize(text)
            punctuations = ['(',')',';',':','[',']','!','"','#','&','*','{','|','}','`','^','_','@','=','<','>','+','$','~','£','..','...'] 
            keywords = [word for word in tokens if  not word in punctuations]
            keywords = " ".join(keywords)
            print(keywords)
            fileOut = open(rootdir+'\\Text\\'+fileName+'.txt',"w", encoding='utf-8')
            fileOut.write(keywords)
            fileOut.close()
```
